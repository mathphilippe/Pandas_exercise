```python
import pandas as pd
import numpy as np
```

### Connect Pandas to the database with this URL: 

```python
titles = pd.read_csv('titles.csv')
titles
```

```python
cast = pd.read_csv('cast.csv')
cast
```

### Using groupby(), count the number of films that have been released each decade in the history of cinema.

```python
bins = [1890, 1899, 1909, 1919, 1929, 1939, 1949, 1959, 1969, 1979, 1989, 1999, 2009, 2019, 2029]
films = titles.groupby(pd.cut(titles['year'], bins)).count()
films.drop('year', axis = 1)
```

### Use groupby() count the number of "Hamlet" films made each decade.

```python
bins = [1890, 1899, 1909, 1919, 1929, 1939, 1949, 1959, 1969, 1979, 1989, 1999, 2009, 2019, 2029]
hamlet = titles[titles['title'] == 'Hamlet'].groupby(pd.cut(titles[titles['title'] == 'Hamlet']['year'], bins)).count()
hamlet.drop('year', axis = 1)
```

### How many leading (n=1) roles were available to actors, and how many to actresses, in each year of the 1950s?

```python
lead = cast[(cast['n'] == 1) & (cast['year'].isin(range(1950, 1960)))].groupby('year').count()
lead.drop(['title', 'name', 'type', 'n'], axis = 1)
```

### In the 1950s decade taken as a whole, how many total roles were available to actors, and how many to actresses, for each "n" number 1 through 5?

```python
actors = cast[(cast['year'].isin(range(1950, 1960))) & (cast['n'].isin(range(1, 6))) & (cast['type'] == 'actor')].groupby('n').count()
actors.drop(['title', 'year', 'name', 'type'], axis = 1)
```

```python
actresses = cast[(cast['year'].isin(range(1950, 1960))) & (cast['n'].isin(range(1, 6))) & (cast['type'] == 'actress')].groupby('n').count()
actresses.drop(['title', 'year', 'name', 'type'], axis = 1)
```

### Use groupby() to determine how many roles are listed for each of the Pink Panther movies.

```python
pinkPanther = cast[cast['title'].str.contains('Pink Panther')].groupby(['title', 'year']).count()
pinkPanther.drop(['name', 'type', 'n'], axis = 1)
```

### List, in order by year, each of the films in which Frank Oz has played more than 1 role.

```python
movies = cast[cast['name'] == 'Frank Oz'].groupby(['year', 'title']).count()
movies[movies['character'] > 1].drop(['name', 'type', 'n'], axis = 1)
```

### List each of the characters that Frank Oz has portrayed at least twice.

```python
character = cast[cast['name'] == 'Frank Oz'].groupby(['character']).count()
character[character['title'] > 1].drop(['year', 'name', 'type', 'n'], axis = 1)
```
